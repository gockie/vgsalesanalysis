```python
tendo_df = pd.read_csv("nintendovgsales.csv")

tendo_df.head()
```

```python
tendo_df.drop(columns = ['Unnamed: 0.1', 'Unnamed: 0', 'Rank'], inplace=True)
```

```python
tendo_df
```

```python
#Filtering by genre
nin_sports_df = tendo_df[tendo_df['Genre'].str.contains('Sports')]
nin_sports_df
```

```python
#Filtering by genre
nin_racing_df = tendo_df[tendo_df['Genre'].str.contains('Racing')]
nin_racing_df
```

```python
#Filtering by genre
nin_action_df = tendo_df[tendo_df['Genre'].str.contains('Action')]
nin_action_df
```

Choosing sports

```python
nin_sports_df.head()
```

```python
#filtering out wii
nin_sports_wii_df = nin_sports_df[nin_sports_df['Platform'].str.contains('Wii')]
nin_sports_wii_df
```

```python
#sorting by top 5 Global_sales
nin_sports_wii_df.sort_values(by='Global_Sales', ascending=False).head()
```

```python
#Plotting to show the top 5 games based on Global sales of Nintendo in the Sports genre on the Wii platform
import matplotlib.pyplot as plt

nin_sports_wii_df.head().plot(x = "Name", y = "Global_Sales", kind = "bar")
plt.title('Top 5 games based on Global sales of Nintendo in the Sports genre on the Wii platform', fontsize = 11)
plt.show()
```

```python
#Filtering/Sorting Global sales by year
nin_sports_wii_GSyear_df = nin_sports_wii_df[['Year', 'Global_Sales']]

nin_sports_wii_GSyear_df
```

```python
meangs = nin_sports_wii_GSyear_df.groupby("Year").mean()
meangs
```

```python
#plotting to show the best performing publishing years for Nintendo in the Sports genre on the Wii platform based on Global sales
meangs.sort_values(by='Global_Sales', ascending=False).plot(kind = "bar" )
plt.title('Best performing publishing years for Nintendo in the Sports genre on the Wii platform', fontsize = 11)
plt.show()
```

```python
#sorting by top 5 North america sales
nin_sports_wii_df.sort_values(by='NA_Sales', ascending=False).head()
```

```python
#Plotting to show the top 5 games based on North America sales of Nintendo in the Sports genre on the Wii platform

nin_sports_wii_df.head().plot(x = "Name", y = "NA_Sales", kind = "bar")
plt.title('Top 5 games based on North America sales of Nintendo in the Sports genre on the Wii platform', fontsize = 11)
plt.show()
```
